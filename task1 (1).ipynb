```python
data=df.set_index('Country Name').iloc[:,4:].T
data=data.apply(pd.to_numeric,errors='coerce').dropna(how='all')
data
```

```python
total=data.sum(axis=1)
plt.figure(figsize=(15,6))
total.plot(kind='bar')
plt.title('Total Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
